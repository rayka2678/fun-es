```
def livro(nome, qtdPagina, autor, preco):
  print(nome, qtdPagina, autor, preco)

def menu():
  nome = str(input('Informe o nome: '))
  qtdPagina = int(input('Informe a quantidade de páginas: '))
  autor= str(input('Informe o nome do autor: '))
  preco = float(input('Informe o preço: '))
  livro(nome, qtdPagina, autor, preco)

menu()
```

```
def alterarNome(nome):
  novoNome = input('Informe o novo nome: ')
  return novoNome

def alterarMatricula(matricula):
  novaMatricula = int(input('Informe sua nova matricula: '))
  return novaMatricula

def aluno():
  nome = input('Informe o seu nome: ')
  matricula = int(input('Informe a sua matricula: '))

  print('\nCadastro realizado:\n ')
  print(f'Nome: {nome}')
  print(f'Matricula: {matricula}\n')

  while True:
    opcao = int(input('Você gostaria de fazer alguma alteração?\n 1 - Sim \n 2 - Não \n'))
    if opcao == 1:
      escolha = int(input('O que você gostaria de alterar?\n 1 - nome \n 2 - matricula\n'))

      if escolha == 1:
        nome = alterarNome(nome)

      elif escolha == 2:
        matricula = alterarMatricula(matricula)
      else:
        print('Opção inválida!\n')

      print('\nCadastro atualizado: ')
      print(f'Nome: {nome}')
      print(f'Matricula: {matricula}\n')

    elif opcao == 2:
      print('Cadastro realizado com sucesso! Sistema finalizado.')
      break

    else:
      print('Resposta inválida, digite apenas o número 1 ou 2!')

aluno()
```

```
def soma(n1, n2, n3):
    return n1 + n2 + n3


def media(n1, n2, n3):
    return soma(n1, n2, n3) / 3


def menu():
    while True:
        try:
            n1 = float(input('Informe o primeiro número: '))
            n2 = float(input('Informe o segundo número: '))
            n3 = float(input('Informe o terceiro número: '))

            total = soma(n1, n2, n3)
            print(f'Soma: {total}')
            print(f'Média: {media(n1, n2, n3):.2f}')

        except ValueError:
            print('Erro! Digite apenas números válidos.')

        repetir = input('Deseja calcular novamente? (s/n): ').strip().lower()
        if repetir != 's':
            print('Programa encerrado.')
            break


menu()
```
